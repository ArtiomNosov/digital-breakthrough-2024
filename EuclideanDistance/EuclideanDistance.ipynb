```python
conn = sqlite3.connect('../parsing/courses.db')
cursor = conn.cursor()
new_cursor = ()
# Execute a query to read data from a table
cursor.execute('SELECT title, skills FROM vacancies')
courses = cursor.fetchall()

def replace_characters(tuple_list, old_char, new_char):
    replaced_tuples = []

    for tup in tuple_list:
        replaced_words = []
        for item in tup:
            new_word = item.replace(old_char, new_char)
            replaced_words.append(new_word)

        replaced_tuples.append(tuple(replaced_words))

    return replaced_tuples
courses = replace_characters(courses, '\n', ' ')
courses
```

```python
f = lambda x: ' '.join(list(x))
courses_text = [f(x) for x in courses]
courses_text
```

```python
scores = []
```

```python
import math

def euclidean_distance(vacancy_skills, course_skills):
    vacancy_skills_set = set(vacancy_skills)
    course_skills_set = set(course_skills)

    common_skills = vacancy_skills_set.intersection(course_skills_set)

    distance = math.sqrt(sum((vacancy_skills.count(skill) - course_skills.count(skill))**2 for skill in common_skills))

    return distance

for i in range(0, len(vacancies_text)):
  for j in range(0, len(courses_text)):
    scores += [(i, j, euclidean_distance(" ".join(list(vacancies_text[i])).split(), " ".join(list(courses_text[j])).split()))]
```
