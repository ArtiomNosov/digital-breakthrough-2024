```python
conn = sqlite3.connect('../parsing/courses.db')
cursor = conn.cursor()
new_cursor = ()
# Execute a query to read data from a table
cursor.execute('SELECT title, skills FROM vacancies')
courses = cursor.fetchall()

def replace_characters(tuple_list, old_char, new_char):
    replaced_tuples = []

    for tup in tuple_list:
        replaced_words = []
        for item in tup:
            new_word = item.replace(old_char, new_char)
            replaced_words.append(new_word)

        replaced_tuples.append(tuple(replaced_words))

    return replaced_tuples
courses = replace_characters(courses, '\n', ' ')
courses
```

```python
f = lambda x: ' '.join(list(x))
courses_text = [f(x) for x in courses]
courses_text
```

```python
scores = []
```

```python

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.metrics.pairwise import cosine_similarity

def cosinSimilarity(resume_words, job_description_words):

  # Преобразование текста в вектора с помощью CountVectorizer и TfidfTransformer
  vectorizer = CountVectorizer()
  transformer = TfidfTransformer()

  resume_vector = vectorizer.fit_transform([resume_words])
  job_description_vector = vectorizer.transform([job_description_words])

  resume_tfidf = transformer.fit_transform(resume_vector)
  job_description_tfidf = transformer.transform(job_description_vector)

  # Вычисление косинусного сходства между векторами
  cosine_sim = cosine_similarity(resume_tfidf, job_description_tfidf)
  return cosine_sim[0][0]

for i in range(0, len(vacancies_text)):
  for j in range(0, len(courses_text)):
    scores += [(i, j, cosinSimilarity(vacancies_text[i], courses_text[j]))]

      
```
