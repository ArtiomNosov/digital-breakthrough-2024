```python
conn = sqlite3.connect('../parsing/courses.db')
cursor = conn.cursor()
new_cursor = ()
# Execute a query to read data from a table
cursor.execute('SELECT title, skills FROM vacancies')
courses = cursor.fetchall()

def replace_characters(tuple_list, old_char, new_char):
    replaced_tuples = []

    for tup in tuple_list:
        replaced_words = []
        for item in tup:
            new_word = item.replace(old_char, new_char)
            replaced_words.append(new_word)

        replaced_tuples.append(tuple(replaced_words))

    return replaced_tuples
courses = replace_characters(courses, '\n', ' ')
courses
```

```python
f = lambda x: ' '.join(list(x))
courses_text = [f(x) for x in courses]
courses_text
```

```python
scores = []
```

```python
import numpy as np

def levenshtein_distance(s, t):
    if s == t:
        return 0
    elif len(s) == 0:
        return len(t)
    elif len(t) == 0:
        return len(s)

    rows = len(s) + 1
    cols = len(t) + 1
    distance = [[0 for _ in range(cols)] for _ in range(rows)]

    for i in range(1, rows):
        distance[i][0] = i
    for j in range(1, cols):
        distance[0][j] = j

    for j in range(1, cols):
        for i in range(1, rows):
            if s[i-1] == t[j-1]:
                cost = 0
            else:
                cost = 1
            distance[i][j] = min(distance[i-1][j] + 1,  # deletion
                                  distance[i][j-1] + 1,  # insertion
                                  distance[i-1][j-1] + cost)  # substitution

    return distance[rows-1][cols-1]

            
for i in range(0, len(vacancies_text)):
  for j in range(0, len(courses_text)):
    scores += [(i, j, levenshtein_distance(" ".join(list(vacancies_text[i])).split(), " ".join(list(courses_text[j])).split()))]
         
```
